```python
import nbformat as nbf
from pathlib import Path

# Change this to the notebook you want to clean
path = Path("TF-IDF Paper.ipynb")

nb = nbf.read(path, as_version=4)
changed = False

for cell in nb.get("cells", []):
    for output in cell.get("outputs", []):
        if "jetTransient" in output:
            del output["jetTransient"]
            changed = True

if changed:
    print("Removed jetTransient from some outputs.")
else:
    print("No jetTransient field found.")

nbf.write(nb, path)
print("Saved cleaned notebook to", path)
```
